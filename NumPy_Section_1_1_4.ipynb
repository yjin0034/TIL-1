# Section 1. N차원 배열 생성

## 1-4. 특정 범위의 값을 가지는 N차원 배열 생성하기

```python
# 리스트의 range()
lst = list(range(0, 10, 3))    # range(start, stop, step)    # 0부터 시작해 10미만까지의 원소를 가지는 리스트 생성하는데, 원소 간 간격을 3으로 설정.
print(lst)
```

```python
# NumPy Array의 arange()
arr = np.arange(9)    # 0부터 시작해 9미만까지의 원소를 가지는 NumPy Array을 생성.(start를 지정하지 않으면, 0부터 시작. step을 지정하지 않으면, 1로 지정)
print(arr)

print()

arr = np.arange(2, 15, 2)    # arange(start, stop, step)    # 2부터 시작해 15미만까지의 원소를 가지는 NumPy Array을 생성하는데, 원소 간 간격을 2로 설정.
print(arr)

print()

arr = np.arange(start=2, stop=15, step=2)    # 인자들의 키워드를 직접 입력해 사용할 수도 있다.
print(arr)
```

```python
# linspace() 함수와 logspace() 함수는 특정 곡선을 그리듯 일정 간격으로 구성된 벡터값을 생성하기 위한 함수입니다.
# linspace : 설정한 범위에서 선형적으로 분할한 위치의 값을 출력
# logspace : 설정한 범위에서 로그로 분할한 위치의 값을 출력


# np.linspace()
# 사용 방법 : np.linspace(a, b, c)
# a와 b사이에 c개의 점(행 벡터)을 생성하기 위한 명령어
# : numpy.linspace(start, end, num=개수, endpoint=True, retstep=False, dtype=자료형)을 사용하여 start ~ end 사이의 값을 개수만큼 생성하여 배열로 반환. 
# (선형 공간에) 생성될 원소의 개수에 따른 각 값의 균등한 등간격으로 분배함.
arr = np.linspace(0, 100, 5)    # 0부터 100까지의 범위 중, 5개의 원소로 균등하게 (선형적으로) 분할한 값들로 구성된 배열 생성. 
print(arr)

print()

arr = np.linspace(0, 100, 250)    # 0부터 100까지의 범위 중, 250개의 원소로 균등하게 (선형적으로) 분할한 값들로 구성된 배열 생성. 
print(arr)
```

```python
# np.logspace() : 로그 스케일(범위, scale) 값으로 배열을 만드는 함수
# a, b, c 을 log scale 등간격인 행 벡터를 생성하는 명령어. log^a와 log^b 사이에 등간격인 c개의 행 벡터를 생성 -> 이해가 잘 안된다. 퍼온 내용인데, 나중에 다시 보고 수정하자. 내가 봤을 땐, 그냥 base로 설정한 값의 거듭제곱 같다?
# : numpy.logspace(start, end, num=개수, endpoint=True, base=10.0, dtype=자료형)을 사용하여 start ~ end 사이의 로그 배율을 사용하여 값을 개수만큼 생성하여 배열로 반환.
# (로그 공간에) 생성될 원소의 개수에 따른 각 값의 균등한 등간격으로 분배함.
arr = np.logspace(1, 10, 10, base=2)    # 1부터 10까지의 범위 중, 10개의 원소로 균등하게 로그 스케일로 분할한 값들로 구성된 배열 생성. base로 설정한 값의 거듭제곱 같다? 2의 거듭제곱    
print(arr)                              # [2.  4.  8.  16.  32.  64.  128.  256.  512.  1024.]
                                        # 2의 거듭제곱  ->  2^1  2^2  2^3  ~  2^9  2^10
print()

arr = np.logspace(1, 10, 10)    # base를 입력하지 않으면, 기본값인 10으로 설정(base=10)되어 상용 로그 스케일(범위, scale)로 균등하게 분배된다. 10의 거듭제곱
print(arr)                      # [1.e+01 1.e+02 1.e+03 1.e+04 1.e+05 1.e+06 1.e+07 1.e+08 1.e+09 1.e+10]
                                # 10의 거듭제곱 -> [10^1  10^2  10^3  ~  10^9  10^10]
print()

arr2 = np.logspace(2, 4, 3)    # 2부터 4까지의 범위 중, 3개의 원소로 균등하게 로그 스케일로 분할한 값들로 구성된 배열 생성. base를 입력하지 않았으므로 상용 로그 스케일로 균등하게 분배
print(arr2)                    # [  100.  1000. 10000.]
                               # 10의 거듭제곱 -> [10^2  10^3  10^4]
```
